```python
# pyramid pattern---


for i in range(1,8):
    print(" " * (8-i), end="")
    ch = ord('A')
    for j in range(i):
        print(chr(ch), end=" ")
        ch += 1
    print(" ")

```

```python
# Reverse Alphabet Traingle

for i in range(1,6):
    ch=ord('A')+ 5-1
    for j in range(i):
        print(chr(ch),end=" ")
        ch-=1
    print()
```

```python
# Diamond alphabet pattern----


ch=ord('A')
for i in range(1,10):
    print(" " *(10-i),end=" ")
    for j in range(i):
        print(chr(ch),end=" ")
    print()
    ch+=1
    
ch-=2
for i in range(10-1,0,-1):
    print(" " *(10-i),end=" ")
    for j in range(i):
        print(chr(ch),end=" ")
    print()
    ch-=1
    
```

```python
#Zig-Zag pattern---

ch = ord('A')
for i in range(5):
    if i % 2 == 0:         
        for j in range(5):
            print(chr(ch), end=" ")
            ch += 1
    else:                   
        temp = ch + 5 - 1
        for j in range(5):
            print(chr(temp), end=" ")
            temp -= 1
        ch += 5
    print()
```

```python
# Number Zig-Zag-----

num=1
for i in range(5):
    if i % 2 == 0:         
        for j in range(5):
            print(num,end=" ")
            num+= 1
    else:                   
        temp = num + 5 - 1
        for j in range(5):
            print(temp,end=" ")
            temp -= 1
        num += 5
    print()
```

```python
#Zig-Zag Right Triangle-----

n = 5
for i in range(1, n+1):
    if i % 2 == 1:
        for j in range(1, i+1):
            print(j, end=" ")
    else:
        for j in range(i, 0, -1):
            print(j, end=" ")
    print()
```
